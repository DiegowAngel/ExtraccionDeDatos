
*FILTROS BÁSICOS PARA VARIABLES CUALITATIVAS*



```python
#Filtro por objeto (1)
filtro1=df[df["continent"] == "South America"]
filtro1

#Aquí selecciono solamente a los países que están en América del Sur
```

```python
#Filtro por filas (2)
filtro2=df.iloc[ 22:28 , : ]  #filas no consecutivas
filtro2

# Se seleccionan de las filas 22 a las 28 solamente
```

```python
#Filtro por columnas (3)
filtro3=df.iloc[ : , [0,1,2,6]]  #columnas no consecutivas
filtro3

# Selecciono solamente las columnas "code", "name", "continent", "population"
```

```python
df= pd.read_csv("CountryTable.csv", index_col=3)
df

#Filtro filas vs  columnas (4)
filtro4=df.loc[["Caribbean","South America"], ["gnp"]]
filtro4

#Selecciono la región de América del Sur y del Caribe, y muestro solamente el producto nacional bruto
```

```python
#Filtro por cabecera (5)
filtro5=df.head(2)
filtro5

#Selcciono solamente los primeros dos elementos
```

*FILTROS BÁSICOS PARA VARIABLES CUANTITATIVAS*


```python
#Filtro por comparacion
filtro6=df[df["gnp"] > 800]
filtro6
#Selecciono a todos los países con un producton nacional bruto mayor a 800
```

```python
#Filtro Y
filtro7=df[(df["gnp"] > 800) & (df["life_expectancy"] >= 70)]
filtro7

#Selcciono todos los países con un gnp mayor a 800 y una expectativa de vida mayor o igual a 70 años
```

```python
#Filtro o
filtro8=df[(df["gnp"] > 800) | (df["life_expectancy"] >= 70)]
filtro8
#Selcciono todos los países con un gnp mayor a 800 o bien, con una expectativa de vida mayor mayor o igual a 70 años
```

```python
#Filtro not
filtro9=df[~(df["gnp"] > 800) & ~(df["life_expectancy"] >= 70)]
filtro9
#Selcciono todos los países con un gnp menor a 800 y con una expectativa de vida menor a 70 años
```
